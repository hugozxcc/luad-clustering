```python
import pandas as pd
import numpy as np
```

```python
DATA_PATH = "../02_data_processed/TCGA_LUAD_norm_logCPM.csv"
```

```python
df = pd.read_csv(DATA_PATH, index_col=0)
```

```python
print("Matrix shape (genes x samples):", df.shape)
print("Any NA:", df.isna().values.any())
```

```python
# Variance-based feature selection
gene_var = df.var(axis=1)

N = 2000
top_genes = gene_var.sort_values(ascending=False).head(N).index
df_top = df.loc[top_genes]

print("Top genes matrix:", df_top.shape)
```

```python
X = df_top.T  # samples x genes
print("Final matrix (samples x genes):", X.shape)

OUT_PATH = "../02_data_processed/TCGA_LUAD_top2000_logCPM_samplesxgenes.csv"
X.to_csv(OUT_PATH)
```
